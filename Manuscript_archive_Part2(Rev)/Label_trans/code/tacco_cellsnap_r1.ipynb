```python
import os
import sys

import pandas as pd
import numpy as np
import anndata as ad

import tacco as tc
#import common_code

from scipy.io import mmread
import scanpy as sc
```

```python
seq = pd.read_csv('../data/tonsil_80k_5khvg_normed.csv')
meta_seq = pd.read_csv('../data/tonsil_80k_meta.csv')
vargene_name = pd.read_csv('../data/5k_vg_names.csv')

xton = pd.read_csv('../../gd_xton/data/Xton_sub_norm.csv')
meta_xton = pd.read_csv('../../gd_xton/data/Xton_sub_meta_snap_feat_clusters.csv')
gene_name = pd.read_csv('../../gd_xton/data/gene_names.csv')
```

```python
xton.columns = gene_name.iloc[:,0].values
```

```python
seq_adata = ad.AnnData(
    seq.T.to_numpy(), dtype=np.float32
)
seq_adata.obs_names = meta_seq['barcode']
seq_adata.var_names = vargene_name.iloc[:,0].values

xton_adata = ad.AnnData(
    xton.to_numpy(), dtype=np.float32
)
xton_adata.var_names = xton.columns
```

```python
# prep xton
xton_adata.obs['x'] = meta_xton['x_centroid']
xton_adata.obs['y'] = meta_xton['y_centroid']
xton_adata.obs['clustername_orig'] = meta_xton['snapres1_ann'].values
xton_adata
```

```python
meta_seq
```

```python
# prep seq
seq_adata.obs.index = meta_seq.loc[:,'barcode'].values
seq_adata.obs['cell'] = seq_adata.obs.index
seq_adata.obs['clustername'] = meta_seq.loc[:,'annotation_level_1'].values
seq_adata.varm['cell'] = pd.DataFrame(seq_adata.X.T.astype(np.float32), index=seq_adata.var.index, columns=seq_adata.obs.index)
seq_adata.varm['clustername'] = pd.DataFrame(seq_adata.X.T.astype(np.float32), index=seq_adata.var.index, columns=seq_adata.obs.index)
```

```python
## get intersect
shared_genes = np.intersect1d(seq_adata.var_names, xton_adata.var_names)
len(shared_genes)
```

```python
seq_adata_share = seq_adata[:,shared_genes]
xton_adata_share = xton_adata[:,shared_genes]
```
