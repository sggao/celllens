```python
## script to tile the stuff
import os
import re
import numpy as np
import pandas as pd
import scipy
from skimage.measure import regionprops
from scipy.io import loadmat
from skimage import io
from skimage.io import imread
from skimage.transform import resize
```

```python
shape_of_each_view=(1008, 1344)
y = shape_of_each_view[0]*9 
x = shape_of_each_view[1]*7
allDataStitch = np.zeros((y,x), dtype = np.float32) # container
```

```python
path_to_segmentation = "/home/bkzhu/super_mario/tonsil_v2/Codex/Images_singleChannel_0503seg/"
path_to_iamges = "/home/bkzhu/super_mario/tonsil_v2/Codex/Images_singleChannel/"
target = "HLA-DR.tif"

all_folders = os.listdir(path_to_iamges)
#all_folders


for x in range(2, 9):
    for y in range(7, 16):
        topleft = [(y-7) * shape_of_each_view[0], (x-2) * shape_of_each_view[1]]
        # get folder name
        x_str, y_str = ('0' + str(x))[-2:], ('0' + str(y))[-2:]
        curr_folder = [name for name in all_folders if re.search('X'+x_str+'_Y'+y_str, name)]
        curr_folder = curr_folder[0]
        
        path_to_iamges_each =  path_to_iamges + curr_folder + '/' + target
        currData = imread(path_to_iamges_each)
        
        allDataStitch[topleft[0]:(topleft[0] + shape_of_each_view[0]),
                      topleft[1]:(topleft[1] + shape_of_each_view[1]) ] = currData
        
io.imsave('/home/bkzhu/super_mario/tonsil_v2/Codex/tiled/HLA-DR_x28_y715.tif',allDataStitch,check_contrast=False)
```
