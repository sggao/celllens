```python
glnsyn = tifffile.imread("/home/bkzhu/spatial_clustering/phase2/data/expresso_brain/2022-08-16T24-25-33_ExPRESSO_brainS3_21-145_A11Gryus_Grey-White_transition_4x12tile/Point1/TIFs/GlnSyn_Simple Segmentation.tiff")

figure(figsize=(12, 12), dpi=120)

#%matplotlib inline

plt.imshow(glnsyn) 
```

```python
glnsyn_label = skimage.measure.label(glnsyn)
glnsyn_label = skimage.morphology.remove_small_objects(glnsyn_label, min_size = 200)
glnsyn_label[glnsyn_label == 1] = 0
figure(figsize=(12, 12), dpi=120)
plt.imshow(glnsyn_label) 
```

```python
len(np.unique(glnsyn_label))
```

```python
# some code for binding morph labels
nuclear_update = nuclear.copy()

for i in np.unique(nuclear):
    print(i)
    if i == 0: # ignore bg
        continue
    temp1 = np.zeros_like(nuclear)
    temp1[nuclear == i] = 1
    temp2 = temp1 * glnsyn_label
    target = np.unique(temp2)
    
    if len(target) > 1: # if contact
        
        temp3 = temp2.copy()
        temp3[temp3!=0] = 1
        
        if np.sum(temp3)>5: # if contact larger than 5 pixels
            print(target)
            
            for j in target:
                if j == 0:
                    continue
                nuclear_update[glnsyn_label == j] = i
```

```python
figure(figsize=(12, 12), dpi=120)
plt.imshow(nuclear_update) 
```
